```python
# configuration json examples can be found in doc
for i in range(1,50,1):
    
    a =  np.random.randint(2,300)
    conf = {
        "type": "regression",
#         "n_classes": np.random.randint(4,100),
        "n_samples": np.random.randint(100,1000),
        "n_features":a,
        "out_path": "./",
        "output": "reg_dataset"+str(i),
#         "n_informative": np.random.random_integers(2,a-1),
        "n_duplicate": np.random.randint(0,int(0.1*a)+1),
        "n_redundant": 0,
#         "n_clusters": 2,
    #     "weights": [0.8, 0.2],
        "pct_missing": 0.30,
        "insert_dollar": "No",
        "insert_percent": "No",
        "n_categorical": 0,
        "star_schema": "No",
        "label_list": [],
#         "n_repeated" : np.random.randint(0,a),       #repeated feature
#         "n_clusters_per_class" : 1 ,
        "noise" : 0.4
        }
    make_dataset(config=conf)
    df = pd.read_csv('reg_dataset'+str(i)+'_train.csv')
    df.rename(columns={"y": "label"},inplace=True)
    df.to_csv("reg_dataset"+str(i)+'_train.csv')
    os.remove("reg_dataset"+str(i)+"_test.csv")
    os.remove("reg_dataset"+str(i)+"_testkey.csv")
    
```
